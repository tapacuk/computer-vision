### Тренування



```python
import torch
from models.custom_cnn import CustomCNN
from utils.dataset import FlowersDataset, get_train_transforms, get_test_transforms
from training.train_baseline import train_baseline
from torch.utils.data import DataLoader

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

train_ds = FlowersDataset("../data/processed/train", transform=get_train_transforms())
val_ds = FlowersDataset("../data/processed/val", transform=get_test_transforms())

train_loader = DataLoader(train_ds, batch_size=32, shuffle=True)
val_loader = DataLoader(val_ds, batch_size=32)

model = CustomCNN(num_classes=5).to(device)

history = train_baseline(
    model,
    train_loader,
    val_loader,
    device,
    epochs=20,
    lr=1e-3,
    patience=5,
    save_path="../checkpoints/baseline_best.pth"
)
```

```python
import matplotlib.pyplot as plt
import os

plt.figure(figsize=(10,5))
plt.plot(history["train_loss"], label="Train Loss")
plt.plot(history["val_loss"], label="Val Loss")
plt.legend()
plt.title("Baseline Loss Curves")

os.makedirs("../results", exist_ok=True)
plt.savefig("../results/baseline_loss.png", dpi=300)
plt.show()

plt.figure(figsize=(10,5))
plt.plot(history["train_acc"], label="Train Accuracy")
plt.plot(history["val_acc"], label="Val Accuracy")
plt.legend()
plt.title("Baseline Accuracy Curves")
plt.savefig("../results/baseline_accuracy.png", dpi=300)
plt.show()
```

перевірка точності тестування

```python
from utils.dataset import get_test_transforms
from torch.utils.data import DataLoader

test_ds = FlowersDataset("../data/processed/test", transform=get_test_transforms())
test_loader = DataLoader(test_ds, batch_size=32)

model.load_state_dict(torch.load("../checkpoints/baseline_best.pth"))
model.eval()

correct = 0
total = 0

with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, pred = torch.max(outputs, 1)
        correct += (pred == labels).sum().item()
        total += labels.size(0)

print("Test Accuracy:", correct / total)
```
